```python
from faker import Faker
import pandas as pd
import random

# Initialize Faker
fake = Faker()

# Create list to store data
data = []

# Generate 200 fake records
for _ in range(10000):
    age = random.randint(18, 65)
    income = round(random.uniform(20000, 120000), 2)
    credit_score = round(random.uniform(300, 850), 2)
    loan_amount = round(random.uniform(1000, 50000), 2)

    # Simple logic: approve if credit score > 600 and income > 30000
    approved = 1 if credit_score > 600 and income > 30000 else 0

    data.append({
        "Age": age,
        "Income": income,
        "Credit_Score": credit_score,
        "Loan_Amount": loan_amount,
        "Approved": approved
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Save as CSV
df.to_csv("credit_dataset.csv", index=False)

print("✅ Dataset generated and saved as 'credit_data.csv'")
```
